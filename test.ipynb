```python
import requests
import os
# LOAD ENVIRONMENT VARIABLES FROM .ENV FILE
import dotenv
dotenv.load_dotenv()
# DISABLE INSECURE REQUEST WARNINGS FROM URLLIB3
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# ---- CONFIGURATION ----
endpointUrl = "https://autoliv-libreoffice-api-endpoint.westeurope.inference.ml.azure.com"

# FETCH THE API KEY FROM THE ENVIRONMENT
apiKey = os.getenv("API_KEY")

headersDict = {"Authorization": f"Bearer {apiKey}"}

# ---- TEST 1: HEALTH CHECK ----
responseHealth = requests.get(f"{endpointUrl}/health", headers=headersDict, verify=False)
print("Health Check:", responseHealth.status_code, responseHealth.json())

# ---- TEST 2: CONVERT A SINGLE FILE ----
# CHANGE TO YOUR TEST FILE PATH
testFile = "INPUT/R200開発歴史-2.2.pptx"  

with open(testFile, "rb") as f:
    responseConvert = requests.post(
        f"{endpointUrl}/convert",
        headers=headersDict,
        files={"file": (os.path.basename(testFile), f)},
        timeout=180,
        verify=False
    )

if responseConvert.status_code == 200:
    outputPath = "OUTPUT/" + os.path.splitext(os.path.basename(testFile))[0] + ".pdf"
    with open(outputPath, "wb") as f:
        f.write(responseConvert.content)
    print(f"✅ Saved: {outputPath} ({len(responseConvert.content)} bytes)")
else:
    print(f"❌ Failed: {responseConvert.status_code} - {responseConvert.text}")
```
